```python
import pandas as pd
from sklearn.ensemble import IsolationForest

df = pd.read_csv("../outputs/risk_scored.csv")

features = df[["magnitude", "depth", "risk_score"]]

iso = IsolationForest(contamination=0.02, random_state=42)
df["anomaly_score"] = iso.fit_predict(features)

df.to_csv("../outputs/anomalies.csv", index=False)

print(df["anomaly_score"].value_counts())

```
