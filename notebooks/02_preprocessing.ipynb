```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("../dataset/indonesia_earthquake.csv")

# ================
# 1. menggubah data
# ================
df ["date"] = pd.to_datetime(df["date"], errors="coerce")

# kombinasi waktu
df["datetime"] = pd.to_datetime(
    df["date"].astype(str) + " " + df["time"], errors="coerce"
)

# ================
# 2. menghapus data yang tidak diperlukan
# ================
df = df.dropna(
    subset=["latitude", "longitude", "depth", "magnitude", ]
)

# ================
# 3. menghapus data duplikat
# ================
df = df.drop_duplicates()

# ================
# 4. membuat fitur frekuensi
# ================
df["location_frequency"] = (
    df.groupby("location")["id"].transform("count")
)

# ================
# 5. menyimpan data yang sudah dibersihkan
# ================
df.to_csv("../outputs/preprocessed.csv", index=False)

print(df.sample)
print(df.head())
```
