```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("../outputs/kde_scored.csv")

print(df.columns.tolist())
scaler = MinMaxScaler()

df["mag_norm"] = scaler.fit_transform(df[["magnitude"]])
df["depth_norm"] = scaler.fit_transform(df[["depth"]])
df["freq_norm"] = scaler.fit_transform(df[["location_frequency"]])
df["density_norm"] = scaler.fit_transform(df[["density_score"]])

# nilai resiko
df["risk_score"] = (
    0.35 * df["mag_norm"]
    + 0.25 * df["freq_norm"]
    + 0.20 * (1-df["depth_norm"])
    + 0.20 * df["density_norm"]
)

df.to_csv("../outputs/risk_scored.csv", index=False)

print(df["risk_score"].describe())
```
